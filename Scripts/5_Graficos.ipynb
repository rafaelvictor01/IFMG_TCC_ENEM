## Evolução do Número de Candidatos por Cluster

## [Classes A vs E] Escolaridade do pai

```python
import numpy as np
import matplotlib.pyplot as plt

anos = ["2019", "2020", "2021", "2022", "2023"]

escolaridade_labels = [
    "Nunca estudou",
    "4ª série Incompleta",
    "8ª série Incompleta",
    "Ensino Médio Incompleto",
    "Ensino Médio Completo",
    "Ensino Superior Completo",
    "Pós-graduação Completa",
    "Não sabe"
]

dados_grupo_A = [
    [13.00, 39.15, 14.47, 8.84, 12.72, 1.49, 0.51, 9.83],
    [12.09, 36.41, 15.95, 9.76, 13.79, 1.21, 0.42, 10.37],
    [8.98, 31.02, 16.38, 11.18, 18.74, 1.97, 0.81, 10.93],
    [5.32, 22.72, 16.42, 13.50, 23.53, 2.66, 1.16, 14.70],
    [6.25, 22.79, 16.32, 13.50, 22.56, 2.32, 1.05, 15.22]
]

dados_grupo_E = [
    [0.30, 2.98, 5.57, 7.65, 32.50, 26.02, 21.58, 3.39],
    [0.16, 1.81, 3.81, 6.02, 28.70, 29.20, 27.73, 2.56],
    [0.12, 1.29, 2.87, 4.72, 23.27, 30.83, 34.19, 2.71],
    [0.11, 1.16, 2.67, 4.52, 22.43, 30.38, 35.91, 2.82],
    [0.10, 1.12, 2.78, 4.54, 21.77, 29.50, 37.03, 3.15]
]

fig, ax = plt.subplots(figsize=(12, 6))

cores = ["#66c2a5", "#fc8d62", "#8da0cb", "#e78ac3", "#a6d854", "#ffd92f", "#e5c494", "#b3b3b3"]

x_pos = np.arange(len(anos))
width = 0.35
espacamento = 0.015  # Pequeno espaçamento entre as barras

for i, ano in enumerate(anos):
    bottom_A, bottom_E = 0, 0
    for j in range(len(escolaridade_labels)):
        ax.bar(x_pos[i] - (width / 2 + espacamento), dados_grupo_A[i][j], width, color=cores[j], bottom=bottom_A)
        bottom_A += dados_grupo_A[i][j]

        ax.bar(x_pos[i] + (width / 2 + espacamento), dados_grupo_E[i][j], width, color=cores[j], bottom=bottom_E)
        bottom_E += dados_grupo_E[i][j]

    ax.text(x_pos[i] - (width / 2 + espacamento), bottom_A + 1, "A", ha="center", va="bottom", fontsize=10, fontweight="semibold", color="gray")
    ax.text(x_pos[i] + (width / 2 + espacamento), bottom_E + 1, "E", ha="center", va="bottom", fontsize=10, fontweight="semibold", color="gray")

handles = [plt.Rectangle((0,0),1,1, color=cores[i]) for i in range(len(escolaridade_labels))]

ax.set_xticks(x_pos)
ax.set_xticklabels(anos)
ax.set_ylabel("Percentual de Candidatos (%)")
ax.set_title("Distribuição da Escolaridade dos Pais - Grupos A e E")

ax.legend(handles, escolaridade_labels, bbox_to_anchor=(1, 1), loc='upper left', ncol=1, frameon=False, fontsize="x-large")

plt.tight_layout()

# Salvando o gráfico
plt.savefig("D:/Graficos/Clusters_A_E_Escolaridade_Pai.png", dpi=300)
plt.show()
```
