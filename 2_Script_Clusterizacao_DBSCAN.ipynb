```python
from sklearn.cluster import DBSCAN
import pandas as pd
```

```python
# ano = '2022'
# ano = '2020'
ano = '2019'

DATASET_ENEM_PATH = 'D:\BASES_PRE_PROCESSADAS\PRE_PROCESSADOS_ENEM_' + ano + '.csv'
```

```python
microdadosEnem = pd.read_csv(DATASET_ENEM_PATH, sep=',', encoding='ISO-8859-1')
```

```python
microdadosEnem.head()
```

```python
# Default eps == 0.5
# Default min_samples == 5


enem_dbscan = DBSCAN(eps=0.5, min_samples=2)
```

```python
enem_dbscan.fit(microdadosEnem)
```

```python
rotulos_clusters = enem_dbscan.labels_
rotulos_clusters
```
